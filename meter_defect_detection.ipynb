### cv2.flip(）函数
flip是opencv的一个图片翻转函数。
```
cv2.flip(src, flipCode[, dst])  # src为要操作的图像
```
| flipCode | Anno         |
| -------- | ------------ |
| 1        | 水平翻转     |
| 0        | 垂直翻转     |
| -1       | 水平垂直翻转 |


### cv2.transpose()
```
cv2.transpose( src[, dst] )
```
参数：
 - src:它是要转置矩阵的图像。
 - dst:它是与src图像大小和深度相同的输出图像。它是一个可选参数。
 - ret:它返回一个图像。

````python
plt.figure()
plt.subplot(1,2,1)
plt.imshow(img) 
plt.subplot(1,2,2)

# 旋转
if rotate=='1':
    img = cv2.flip(cv2.transpose(img), 1)
elif rotate=='2':
    img=cv2.flip(img, -1)
elif rotate=='3':
    img= cv2.flip(cv2.transpose(cv2.flip(img, -1)), 1)
# 灰度化   
if len(img.shape)==3:
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
else:
    gray=img.copy()
plt.imshow(img) 
plt.show()

plt.figure()
plt.imshow(gray)
````

````python
w1, h1 = img.shape[:2]
center = (h1 // 2, w1 // 2)
w1, h1, center
````

````python
if h2>700:
    l=0.2
elif h2>500:
    l=0.5
elif h2>300:
    l=0.7
else:
    l=0.8
h2, l
````

createLineSegmentDetector 的第一个参数取值有：
 - LSD_REFINE_NONE：没有改良的方式；
 - LSD_REFINE_STD： 标准改良方式，将带弧度的线（拱线）拆成多个可以逼近原线段的直线度；
 - LSD_REFINE_ADV： 进一步改良方式，计算出错误警告数量，通过增加精度，减少尺寸进一步精确直线。

````python
lsd = cv2.createLineSegmentDetector(cv2.LSD_REFINE_NONE, l) # 不缩放
````

````python
lines = lsd.detect(gray)
print(len(lines))
print(len(lines[0]))
````

### cv2.line()-画线条
参数：
 - img：      要划的线所在的图像;
 - pt1：      直线起点
 - pt2：      直线终点  （坐标分别为宽、高,opencv中图像的坐标原点在左上角）
 - color：     直线的颜色
 - thickness=1： 线条粗细,默认是1.如果一个闭合图形设置为-1，那么整个图形就会被填充。

````python
# LSD算法检测结果
img_lsd = img.copy()
for dline in lines[0]:
    x0 = int(round(dline[0][0]))     
    y0 = int(round(dline[0][1]))    
    x1 = int(round(dline[0][2]))     
    y1 = int(round(dline[0][3]))     
    cv2.line(img_lsd, (x0, y0), (x1,y1), 255, 1, cv2.LINE_AA) 
plt.figure() 
plt.imshow(img_lsd)
````

 ### 相关参考链接

````python
for line in lines[0]:
    '''
    x1:line[0][0]
    y1:line[0][1]
    x2:line[0][3]
    y2:line[0][4]
    '''
    if np.abs(line[0][0] - line[0][2])>(4*w2/5) and np.abs(line[0][1] - line[0][3])<(h2/10): # 先分离出长度和模板差不多的横向直线
        line2=line
        x0 = int(round(line2[0][0]))     
        y0 = int(round(line2[0][1]))    
        x1 = int(round(line2[0][2]))     
        y1 = int(round(line2[0][3]))     
        cv2.line(img, (x0, y0), (x1,y1), 255, 1, cv2.LINE_AA) 
plt.figure() 
plt.imshow(img)
````

### 直线的斜率表达式
$$
angle = \frac{180}{\pi}  arctan (\frac{y2-y1}{x2-x1}) 
$$
### cv2.getRotationMatrix2D(center,angle,scale)
用于构建旋转矩阵matrix
 - center: 图片的旋转中心
 - angle: 旋转角度
 - scale: 旋转后图像相比原来的缩放比例
 - 返回值: 计算得到的旋转矩阵
 
### cv2.warpAffine(img,matrix,size)
仿射变换
 - img: 输入图像
 - matrix: 由cv2.getRotationMatrix2D()获得
 - size: 输出图像的大小，格式为（宽(列），高（行）)

````python
'''
获取直线的旋转角度，通过getRotationMatrix2D获取旋转角度对图像进行矫正
以达到与模板图片相同的位置
'''
angle = np.arctan((float)(line2[0][3] - line2[0][1]) /
                          (float)(line2[0][2] - line2[0][0])) * 180 / np.pi
M = cv2.getRotationMatrix2D(center, angle, 1.0) 
dst_img = cv2.warpAffine(img, M, (h1, w1))
# print(M)
# print(rotated)
plt.figure() 
plt.imshow(dst_img) # 旋转仿射之后的图像
````

````python
# 灰度化
if len(dst_img.shape)==3:
    gray2 = cv2.cvtColor(dst_img, cv2.COLOR_BGR2GRAY)
else:
    gray2=dst_img.copy()
plt.figure() 
plt.imshow(gray2) # 旋转仿射之后的图像
````

### cv2.matchTemplate(image, templ, method, result=None, mask=None)

opencv模板匹配

 - image：待搜索图像
 - templ：模板图像
 - result：匹配结果
 - method：计算匹配程度的方法
 
关于参数 **method**：
 - CV_TM_SQDIFF 平方差匹配法：该方法采用平方差来进行匹配；最好的匹配值为0；匹配越差，匹配值越大。
 - CV_TM_CCORR 相关匹配法：该方法采用乘法操作；数值越大表明匹配程度越好。
 - CV_TM_CCOEFF 相关系数匹配法：1表示完美的匹配；-1表示最差的匹配。
 - CV_TM_SQDIFF_NORMED 归一化平方差匹配法
 - CV_TM_CCORR_NORMED 归一化相关匹配法
 - CV_TM_CCOEFF_NORMED 归一化相关系数匹配法
 
### cv2.minMaxLoc(src, mask=None)

src是一个矩阵。

函数功能：假设有一个矩阵a,现在需要求这个矩阵的最小值，最大值，并得到最大值，最小值的索引。咋一看感觉很复杂，但使用这个cv2.minMaxLoc()函数就可全部解决。函数返回的四个值就是上述所要得到的。

````python
src = cv2.matchTemplate(gray2, template, cv2.TM_CCOEFF_NORMED)
mn_val, max_val, min_loc, max_loc = cv2.minMaxLoc(src)
mn_val, max_val, min_loc, max_loc
````

````python
top_left2 = max_loc
bottom_right2 = (top_left2[0] + w2, top_left2[1] + h2)
roi = dst_img[top_left2[1]:bottom_right2[1], top_left2[0]:bottom_right2[0]] # 匹配出结果在原图中的位置
top_left2, bottom_right2
````

### numpy.std(a, axis=None, dtype=None, out=None, ddof=0)

功能：计算沿指定轴的标准差。返回数组元素的标准差
 - a： array_like，需计算标准差的数组
 - axis： int, 可选，计算标准差的轴。默认情况是计算扁平数组的标准偏差。
   axis=0时，表示求每一列标准差，axis=1时，表示求每一行标准差，当axis=None时，表示求全局标准差。
 - dtype： dtype, 可选，用于计算标准差的类型。对于整数类型的数组，缺省值为Float 64，对于浮点数类型的数组，它与数组类型相同。
 - out： ndarray, 可选，将结果放置在其中的替代输出数组。它必须具有与预期输出相同的形状，但如果有必要，类型(计算值的类型)将被转换。
 - ddof： int, 可选，Delta的自由度
